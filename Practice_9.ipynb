```python
import pandas as pd
import numpy as np
import os
from sklearn.model_selection import train_test_split

cancer = pd.read_csv('breast_cancer_dataset.csv')
display(cancer)
```

```python
# switch ones and zeroes
cancer = cancer.drop(columns = ['class'])
cancer['class'] = [1 if x == 'Malign' else 0 for x in cancer['class_names']]
cancer = cancer.drop(columns = ['class_names'])
cancer
```

```python
# separate into features and labels
features = cancer.drop(columns = ['class'])
labels = cancer['class']

# split into test and train
X_train, X_test, y_train, y_test = train_test_split(features, labels, random_state=0)
```

```python
# creating prepocessing pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer

pre_pipeline = Pipeline([('imputer', SimpleImputer(strategy='median')), ('scaler', StandardScaler())])
```

```python
# knn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import recall_score


knn_pipeline = Pipeline([('prepare', pre_pipeline), ('knn', KNeighborsClassifier())])

param_grid_knn = {
    'knn__n_neighbors': np.arange(1, 8)
}

grid_search_knn = GridSearchCV(knn_pipeline, param_grid_knn, cv=5)
grid_search_knn.fit(X_train, y_train)
best_knn = grid_search_knn.best_estimator_

print('KNN:\n')
print('Best parameters: {}'.format(grid_search_knn.best_params_))
print('\nRecall Score:\n')
print('Training: {}'.format(recall_score(y_train, best_knn.predict(X_train))))
print('Test: {}\n'.format(recall_score(y_test, best_knn.predict(X_test))))
```

```python
# log reg
from sklearn.linear_model import LogisticRegression

logreg_pipeline = Pipeline([('prepare', pre_pipeline), ('lr', LogisticRegression())])

param_grid_lr = {
    'lr__C': [0.001, 0.01, 0.1, 1, 10],
    'lr__class_weight': [None, 'balanced']
}

grid_search_lr = GridSearchCV(logreg_pipeline, param_grid_lr, cv=5)
grid_search_lr.fit(X_train, y_train)
best_lr = grid_search_lr.best_estimator_

print('Log Reg:\n')
print('Best parameters: {}'.format(grid_search_lr.best_params_))
print('\nRecall Score:\n')
print('Training: {}'.format(recall_score(y_train, best_lr.predict(X_train))))
print('Test: {}\n'.format(recall_score(y_test, best_lr.predict(X_test))))
```

```python
#decision tree
from sklearn.tree import DecisionTreeClassifier

tree_pipeline = Pipeline([('prepare', pre_pipeline), ('dt', DecisionTreeClassifier())])

param_grid_tree = {
    'dt__max_depth': np.arange(2, 10),
    'dt__max_features': np.arange(2, 10)
}

grid_search_tree = GridSearchCV(tree_pipeline, param_grid_tree, cv=5)
grid_search_tree.fit(X_train, y_train)
best_tree = grid_search_tree.best_estimator_

print('Decision Tree:\n')
print('Best parameters: {}'.format(grid_search_tree.best_params_))
print('\nRecall Score:\n')
print('Training: {}'.format(recall_score(y_train, best_tree.predict(X_train))))
print('Test: {}\n'.format(recall_score(y_test, best_tree.predict(X_test))))
```

```python
# random forest
from sklearn.ensemble import RandomForestClassifier

rf_pipeline = Pipeline([('prepare', pre_pipeline), ('rf', RandomForestClassifier())])

param_grid_rf = {
             'rf__max_depth': [4, 5, 6],
             'rf__n_estimators' : [100, 200],
             'rf__max_features': [9, 10, 12, 15],
}

grid_search_rf = GridSearchCV(rf_pipeline, param_grid_rf, cv=5)
grid_search_rf.fit(X_train, y_train)
best_rf = grid_search_rf.best_estimator_

print('Random Forest:\n')
print('Best parameters: {}'.format(grid_search_rf.best_params_))
print('\nRecall Score:\n')
print('Training: {}'.format(recall_score(y_train, best_rf.predict(X_train))))
print('Test: {}\n'.format(recall_score(y_test, best_rf.predict(X_test))))
```

```python
from sklearn.svm import SVC

svc_pipeline = Pipeline([('prepare', pre_pipeline), ('svc', SVC())])

param_grid_svc = [
    {
        'svc__kernel': ['poly'],
        'svc__C': [0.1, 1, 10, 100],
        'svc__degree': [1, 2, 3, 4, 5, 6]
    },
    {
        'svc__kernel': ['rbf'],
        'svc__C': [0.1, 1, 10, 100],
        'svc__gamma': [.1, .5, .9]
    }
]

grid_search_svc = GridSearchCV(svc_pipeline, param_grid_svc, cv=5)
grid_search_svc.fit(X_train, y_train)
best_svc = grid_search_svc.best_estimator_

print('SVC:\n')
print('Best parameters: {}'.format(grid_search_svc.best_params_))
print('\nRecall Score:\n')
print('Training: {}'.format(recall_score(y_train, best_svc.predict(X_train))))
print('Test: {}\n'.format(recall_score(y_test, best_svc.predict(X_test))))
```

```python
# linear svc
from sklearn.svm import LinearSVC

linsvc_pipeline = Pipeline([('prepare', pre_pipeline), ('linsvc', LinearSVC())])

param_grid_linsvc = {
    'linsvc__C': [0.001, 0.01, 0.1, 1, 10, 100]
}

grid_search_linsvc = GridSearchCV(linsvc_pipeline, param_grid_linsvc, cv=5)
grid_search_linsvc.fit(X_train, y_train)
best_linsvc = grid_search_linsvc.best_estimator_

print('Linear SVC:\n')
print('Best parameters: {}'.format(grid_search_linsvc.best_params_))
print('\nRecall Score:\n')
print('Training: {}'.format(recall_score(y_train, best_linsvc.predict(X_train))))
print('Test: {}\n'.format(recall_score(y_test, best_linsvc.predict(X_test))))
```

```python
from sklearn.ensemble import VotingClassifier

best_lr
best_rf
best_svc


voting_clf = VotingClassifier(estimators=[('lr', best_lr), ('rf', best_rf), ('svc', best_svc)],voting='hard')
voting_clf.fit(X_train, y_train)
```
